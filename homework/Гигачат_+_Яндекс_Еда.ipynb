# Описание

Необходимо на основе Гигачата создать сервис, который будет на запрос блюда будет создать рецепт, а также предлагать необходимые продукты в Купере (бывш. Сбермаркет). Результатом работы будет текст в формате Markdown.

## Последовательность выполнения

1. Знакомитесь с [API Гигачата](https://developers.sber.ru/docs/ru/gigachat/api/reference/rest/gigachat-api)
2. Генерируете токен для работы с API
3. Прочитайте [про промты и рекомендации по их созданию](https://developers.sber.ru/docs/ru/gigachat/prompt-design)
4. Составьте промт, который на ваш запрос будет выводить рецепт блюда в формате json. Пример ответа в формате json на запрос "Борщ":
```json
{
  "recipe": {
    "title": "Классический борщ",
    "description": "Традиционный русский суп с насыщенным вкусом.",
    "ingredients": [
      {"name": "говядина на кости", "amount": "500 г"},
      {"name": "свекла", "amount": "2 шт."},
      {"name": "картофель", "amount": "4 шт."},
      {"name": "морковь", "amount": "1 шт."},
      {"name": "лук репчатый", "amount": "1 шт."},
      {"name": "капуста белокочанная", "amount": "300 г"},
      {"name": "томатная паста", "amount": "2 ст.л."},
      {"name": "чеснок", "amount": "2 зубчика"},
      {"name": "лавровый лист", "amount": "2 шт."},
      {"name": "соль", "amount": "по вкусу"},
      {"name": "перец черный молотый", "amount": "по вкусу"},
      {"name": "сахар", "amount": "1 ч.л."},
      {"name": "уксус яблочный", "amount": "1 ст.л."},
      {"name": "растительное масло", "amount": "2 ст.л."}
    ],
    "instructions": [
      "Поставь говядину вариться на медленном огне в большой кастрюле с водой. Варить около 1,5 часов до готовности мяса.",
      "Свеклу очисти и натри на крупной терке. Обжарь свеклу на растительном масле с добавлением уксуса и сахара, затем добавь томатную пасту и туши еще 10 минут.",
      "Морковь и лук очисти и мелко нарежь. Обжарь их на сковороде до золотистого цвета.",
      "Картофель очисти и нарежь кубиками. Добавь картофель в бульон после того, как мясо сварится.",
      "Капусту нашинкуй и добавь в кастрюлю через 10 минут после картофеля.",
      "Добавь обжаренную свеклу, морковь и лук в кастрюлю. Посоли и поперчи по вкусу. Добавь лавровый лист и чеснок.",
      "Варить еще 15-20 минут до готовности всех ингредиентов."
    ]
  }
}
```
5. Найдите необходимые продукты в Яндекс.Еде. Для этого вам может пригодиться ручка `api/v1/menu/search`:

```bash
curl --location 'https://eda.yandex.ru/api/v1/menu/search' \
--header 'accept: application/json, text/plain, */*' \
--header 'accept-language: ru' \
--header 'content-type: application/json;charset=UTF-8' \
--data '{
    "region_id": 1,
    "place_slug": "perekrestok",
    "text": "молоко",
    "location": {
        "lat": 55.7106035,
        "lon": 37.743341
    }
}'
```
6. В ответе ручки выбрать подходящие вам продукты, для каждого продукта есть поле public_id (например, `f9973129-ca8e-4fca-b922-9e44ca07d2f8`). Готовый url должен выглядеть сл. образом: https://eda.yandex.ru/retail/perekrestok?item=f9973129-ca8e-4fca-b922-9e44ca07d2f8. То есть https://eda.yandex.ru/retail/perekrestok?item=[public_id].
7. Подготовить рецепт к читаемому виду. Предлагается привести ответ к формату Markdown (это формат текста в Jupyter ноутбуке, например). Пример готового рецепта будет ниже.
8. Подготовить отчёт о проделанной работе (тоже будет ниже).
